```python
import random
import numpy as np
import matplotlib.pyplot as plt
from numpy.linalg import pinv
import math
from numpy.linalg import inv

x = []
y = []
#a,b
for i in range(50):
    x_temp = i + 1
    u = np.random.uniform(-1, 1)
    y_temp = i + 1 + u
    x.append(x_temp)
    y.append(y_temp)
    
M = [[1 for i in range(0,50)],x]
#c
W = np.matmul(np.array(y), pinv(np.array(M)))
Y = [W[0] + W[1]*x[i] for i in range(0,50)]
#d
yn = np.polyval([W[1], W[0]], np.array(x))
fig, ax = plt.subplots(figsize=(10,10))
plt.scatter(x, y, c = 'red')
plt.plot(x, yn, c = 'blue')
plt.ylabel('Y - values')
plt.xlabel('X - values')
plt.title('Linear Least Squares')
plt.show()
```

```python
w=[0.9,0.25]

xs=sum(x)
ys=sum(y)

X = np.array(x)
Y = np.array(Y)
alpha = 0.00001/2
ep=0
#Gradient Descent 
while(ep<5):
    ep+=1
    Xgrad=0
    Ygrad=0
    for i in range(50):
        Xgrad = Xgrad+(-2)*(y[i]-(w[0]+w[1]*x[i]))
        Ygrad = Ygrad + (-2)*(y[i]-(w[0]+w[1]*x[i]))*x[i]
    gradupdate = np.array([Xgrad, Ygrad])
    update = - alpha * gradupdate
    if np.linalg.norm(w - np.add(w,update)) < 0.0000000001:
        break
    else:
        w = np.add(w,update)
    
print (w[0], w[1])
```

```python
yn = np.polyval([w[0], w[1] ], np.array(x))
fig, ax = plt.subplots(figsize=(10,10))
plt.scatter(x, y, c = 'red')
plt.plot(x , yn, c = 'blue')
plt.ylabel('Y - values')
plt.xlabel('X - values')
plt.title('Grad Descent')
plt.show()
```
