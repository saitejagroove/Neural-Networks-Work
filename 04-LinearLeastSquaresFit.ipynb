```python
import random
import numpy as np
import matplotlib.pyplot as plt
from numpy.linalg import pinv

x = [i+1 for i in range(0,50)]
y = [i + 1 + random.uniform(-1,1) for i in range(0,50)]
M = [[1 for i in range(0,50)],x]
W = np.matmul(np.array(y), pinv(np.array(M)))
Y = [W[0] + W[1]*x[i] for i in range(0,50)]
plt.title("Points (x,y) with their linear least squares fit\n",fontsize=16)
plt.xlabel('X-Values',fontsize=14)
plt.ylabel('Y-Values',fontsize=14)
plt.scatter(x, y, color = 'red', marker='s', label='Points (x,y)')
plt.plot(x, Y, linewidth=3, color='black', label='Linear least squares fit')
plt.legend(bbox_to_anchor=(1,1), loc='best', title="Legends\n")
plt.show()
```
